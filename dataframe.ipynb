```python
import pandas as pd
import numpy as np
```

```python
divorce = pd.read_csv('divorce.csv')
```

```python
divorce['year']
divorce.rename(columns={'HS_3544': 'HS'}, inplace=True)

# Loop through column names and update to delete the characters after the underscore
for column in divorce.columns:
    
    if '_' in column:
        new_column_name = column.split('_')[0]
        divorce.rename(columns={column: new_column_name}, inplace=True)


```

```python
divorce.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
divorce['date'] = pd.to_datetime(divorce['date'])
```

```python
divorce = divorce.loc[:, ~divorce.columns.duplicated()]
divorce
```

```python

# Fill NaN with the mean of the column named GD
divorce['GD'].fillna(divorce['GD'].mean(), inplace=True)
divorce
```

```python
divorce.columns.str.upper()
divorce
```

```python
column_name_mapping = {
    old_name: old_name.upper()
      for old_name in divorce.columns
    }
column_name_mapping
```

```python
divorce.rename(columns=column_name_mapping, inplace=True)
divorce
```

```python
divorce.to_csv('./Results/divorce_cleaned.csv', index=False)
divorce['INDEX'] = range(1, len(divorce) + 1)
divorce = divorce[['INDEX'] + [col for col in divorce.columns if col != 'INDEX']]
```
